```python
#Exploring names of all cities

df['City'].unique()
```

```python
#Exploring which cities have the higgest number of orders

df['City'].value_counts(ascending = False)
```

```python
#Exploring which cities have the lowest number of orders

df['City'].value_counts(ascending = True)
```

```python
#Each state and their Sales

plt.figure(figsize=(20,5))
plt.title("Each state and their Sales")
sns.barplot(x = df['State'] , y = df["Sales"])
plt.xticks(rotation=90)
```

```python
#Each state and their profit

plt.figure(figsize=(20,5))
plt.title("Each state and their profit")
sns.barplot(x = df['State'] , y = df["Profit"])
plt.xticks(rotation=90)
```

```python
# Correlation between Sales, Quantity, Discount and Profit

plt.figure(figsize=(20,5))
sns.heatmap(df.corr(), cmap = 'Reds', annot = True)
plt.show()
```
